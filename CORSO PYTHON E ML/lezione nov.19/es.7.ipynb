```python
import pandas as pd

data = {
'Regione': ['Nord', 'Sud', 'Nord', 'Est', 'Ovest', 'Sud', 'Est', 'Ovest', 'Sud', 'Est'],
'Prodotto': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B','A', 'B'],
'Vendite': [100, 150, 200, 250, 300, 350, 400, 450,311,385]
}

df=pd.DataFrame(data)
df
```

```python
#mostrare la media delle vendite di ogni combinazione di Regione e Prodotto
df1=df.pivot_table(index='Regione',values='Vendite',columns='Prodotto',aggfunc='mean')
df1
```

```python
#creare funione che calcola varianza

def varianza(x):
    return x.var()

df2=df.pivot_table(index='Regione',values='Vendite',columns='Prodotto',aggfunc=varianza,fill_value=0)
df2
```
