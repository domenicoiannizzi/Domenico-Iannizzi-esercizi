GRAFICO CON X E Y

```python
x=np.linspace(0,2*np.pi,10)
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.title('Sine wave')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

SCATTER PLOT EXAMPLE

```python
plt.scatter(x,y)
```

BAR CHART EXAMPLE


```python
categories=['A','B','C','D']
values=[10,20,30,40]

plt.bar(categories, values, color=['red', 'blue', 'green', 'purple'])
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

PIE CHART EXAMPLE

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Programming Language Popularity")
plt.axis('equal')
plt.show()
```

PEAK ANNOTATION EXAMPLE

```python
plt.plot(x, y, label='Sine Wave')
plt.title("Annotated Sine Wave")
plt.xlabel("x values")
plt.ylabel("sin(x)")
plt.legend()

peak_x = np.pi / 2
peak_y = 1
plt.annotate('Peak', xy=(peak_x, peak_y), xytext=(peak_x + 1, peak_y - 0.5),
arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

DUAL AXIS EXAMPLE

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('x values')
ax1.set_ylabel('sin(x)', color=color)
ax1.plot(x, y, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('cos(x)', color=color)
ax2.plot(x, np.cos(x), color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title("Dual Axes Example")
fig.tight_layout() # To prevent labels from overlapping
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# First subplot: Histogram
axes[0, 0].hist(np.random.standard_normal(100), bins=20, color="black", alpha=0.3)
axes[0, 0].set_title("Histogram")

# Second subplot: Scatter Plot
axes[0, 1].scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))
axes[0, 1].set_title("Scatter Plot")

# Third subplot: Line Plot
axes[1, 0].plot(np.random.standard_normal(50).cumsum(), color="black", linestyle="dashed")
axes[1, 0].set_title("Cumulative Sum Line Plot")

# Fourth subplot: Empty or another plot
axes[1, 1].text(0.5, 0.5, 'Empty Plot', horizontalalignment='center', verticalalignment='center')
axes[1, 1].set_title("Empty")

plt.tight_layout()
plt.show()
```
