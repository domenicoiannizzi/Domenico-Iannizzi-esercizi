```python
import numpy as np
import pandas as pd

data = {
'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack', 'Kara', 'Liam'],
'Department': ['HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT'],
'Salary': [50000, 60000, 70000, 52000, 58000, 72000, 55000, 62000, 75000, 51000, 63000, 76000],
'Experience': [2, 5, 8, 3, 4, 10, 4, 6, 12, 1, 7, 15]
}

df = pd.DataFrame(data)
df.head()
```

```python
#la media dello stipendio e dell'esperienza per ogni dipartimento

media_stip_esp_dipartimento = df.groupby('Department').mean('Experience', 'Salary')

media_stip_esp_dipartimento
```

```python
#trovare i dipendendi con uno stipendio sopra la media del loro dipartimento

salary_medio=df.groupby('Department')['Salary'].transform('mean')

df1 = df[df['Salary'] > salary_medio]

df2=df1.sort_values(by='Department', ascending=True)

df2

```

```python
#aggiungere una nuova colonna che rappresenta il rank (consiglio cercate la funzione rank di pandas) di
#  ogni dipendente all'interno del suo dipartimento

df['Rank'] = df.groupby('Department')['Salary'].rank(ascending=False)
print(df)
```

```python
#Mostrare il fataframe raggruppato per dipartimento e ogni dipendente in ordine dal rank più alto al più basso

df2 = df.sort_values(['Department', 'Rank'], ascending=[True, False])

df2
```
