```python
import pandas as pd
import numpy as np
data = {
    "Nome": ["Alice", "Bob", "Carla", "Davide", "Elena", "Fabio", "Gina", "Hassan", "Irene", "Luca", 
             "Marta", "Nina", "Oscar", "Paola", "Quentin", "Rosa", "Simone", "Tina", "Umberto", "Vera"],
    "Età": [25, 34, 29, 45, 31, 28, 39, 40, 26, 50, 27, 33, 41, 24, 37, 43, 30, 32, 38, 36],
    "Città": ["Milano", "Roma", "Reggio Calabria", "Torino", "Genova", "Bologna", "Firenze", "Palermo", "Verona", "Venezia",
              "Cagliari", "Parma", "Trieste", "Ancona", "Perugia", "Trento", "Rimini", "Pisa", "Siena", "Modena"],
    "Salario": [2500, 3400, 2900, 4500, 3100, 2800, 3900, 4000, 2600, 5000, 
                2700, 3300, 4100, 2400, 3700, 4300, 3000, 3200, 3800, 3600]
}

x=pd.DataFrame(data)
df=x.sample(n=10)
df
```

```python
df["Salario"].fillna(df["Salario"].median(), inplace=True)
```

```python
def livello_eta(eta):
    if eta <= 18:
        return "Giovane"
    elif 19 <= eta <= 65:
        return "Adulto"
    else:
        return "Senior"
    
df["Età_Categoria"] = df["Età"].apply(livello_eta)
df
```

```python
df.to_csv("dataframe.csv")
```
